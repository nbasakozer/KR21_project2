```python
from BayesNet import BayesNet
from BNReasoner import BNReasoner
import pandas as pd
```

### Lecture Example 2

```python
# load bayesian network and initialize reasoner
bn_le2 = BayesNet()
bn_le2.load_from_bifxml('testing/lecture_example2.BIFXML')
bnr_le2 = BNReasoner(bn_le2)
```

```python
# visualize problem
bn_le2.draw_structure()
```

```python
# display variables
bn_le2.get_all_variables()
```

```python
# display conditional probability tables
bn_le2.get_all_cpts()
```

### Network Pruning

```python
pruned_nw = bnr_le2.pruning(['I'],{'J': True, 'O': False})
pruned_nw.draw_structure() # correct - after 347234 years!
pruned_nw.get_all_cpts()
```
